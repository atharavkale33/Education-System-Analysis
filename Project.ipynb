# MATPLOTLIB (Basic Visualization)

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

#  # . Bar Chart

```python
df['Course'].value_counts().plot(kind='bar')
plt.title("Students per Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df['Age'], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x=df['Final_Exam_Marks'])
plt.title("Marks Distribution")
plt.show()
```

```python
plt.scatter(df['Study_Hours_Per_Week'], df['Final_Exam_Marks'])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

```python
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

The dataset shows variation in student performance across different subjects.
Students who study more hours generally score higher marks.
Average marks indicate a moderate overall academic performance.
A few students score significantly above the class average (top performers).
Some students score below average, indicating need for extra support.
Marks distribution is slightly skewed towards higher values.
Consistent study habits positively impact student results.
There is a clear relationship between effort (study hours) and outcome (marks).
Performance differences suggest varied learning abilities among students.
